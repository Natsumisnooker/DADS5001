# ต้องการให้ Cleansing Data ใน Colum :
1. Company name ตัดตัวเลข และ อักษรพิเศษ ออกจากข้อความ
2. Location แยกข้อความหลัง comma โดยสร้างเป็นคอลัมน์ใหม่ ใช้ชื่อว่า State
3. ต้องการให้แปลงค่าในคอลัมน์ ต่อไปนี้ ( Headquarters Size Founded Type of ownership Industry Sector Revenue Salary Estimate Rating โดยมีค่าเป็น -1 ให้แปลงเป็น N/A ต้องการให้แยกคอลัมน์ Salary Estimate เป็น สองคอลัมน์ แยกด้วยเครื่อง '-' และลบ (Glassdoor est.) และ $ และ ปรับ K เป็น 000 และ ก่อน sperate และ drop colum : salary estimate
เขียนโค้ชบนColabและ export ไฟล์ ผลลัพธ์เป็น Xls

```python
import pandas as pd
import re
from google.colab import files

# STEP 1: Upload file
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# STEP 2: Read file (CSV/XLSX)
if file_name.endswith(".csv"):
    df = pd.read_csv(file_name)
elif file_name.endswith(".xlsx"):
    df = pd.read_excel(file_name)
else:
    raise ValueError("⚠️ รองรับเฉพาะ CSV และ Excel เท่านั้น")

print("✅ โหลดไฟล์สำเร็จ")
print("คอลัมน์ที่พบ:", list(df.columns))

# STEP 3: Split number from Company Name
#if "Company Name" in df.columns:
    # df["Number"] = df["Company Name"].str.extract(r'(\d+)$')   # ตัวเลขท้ายสุดหลัง space
    # df["Company Name"] = df["Company Name"].str.replace(r'\d+$', '', regex=True).str.strip()
    # ลบตัวเลข + จุดทศนิยม 3 หลักสุดท้ายออกจาก Company Name
    # ดึงตัวเลขทศนิยม 3 หลักสุดท้ายมาเก็บใน Number
    # df["Number"] = df["Company Name"].str.extract(r'(\d+\.\d{3})$')
    # df["Company Name"] = df["Company Name"].str.replace(r'\d+\.\d{3}$', '', regex=True).str.strip()

# STEP 4: Extract State from Location
if "Location" in df.columns:
    df["State"] = df["Location"].apply(
        lambda x: str(x).split(",")[-1].strip() if pd.notna(x) and "," in str(x) else ""
    )

# STEP 5: Replace -1 with N/A
cols_to_replace = [
    "Headquarters","Size","Founded","Type of ownership",
    "Industry","Sector","Revenue","Salary Estimate","Rating"
]
for col in cols_to_replace:
    if col in df.columns:
        df[col] = df[col].replace(-1, "N/A").replace("-1", "N/A")

# STEP 6: Clean & Split Salary Estimate
if "Salary Estimate" in df.columns:
    def clean_salary(s):
        if pd.isna(s) or s in ["-1", -1, "N/A"]:
            return "N/A"
        s = str(s)
        s = s.replace("(Glassdoor est.)", "").replace("$", "").strip()
        s = re.sub(r'(\d+)K', lambda m: str(int(m.group(1)) * 1000), s)  # 50K → 50000
        return s.strip()

    df["Salary Estimate Clean"] = df["Salary Estimate"].apply(clean_salary)

    salary_split = df["Salary Estimate Clean"].str.split("-", n=1, expand=True)
    df["Salary Min"] = salary_split[0].str.strip().replace("", "N/A")
    df["Salary Max"] = salary_split[1].str.strip().replace("", "N/A") if salary_split.shape[1] > 1 else "N/A"

    df = df.drop(columns=["Salary Estimate", "Salary Estimate Clean"])

# STEP 7: Preview
print("\n--- ตัวอย่างข้อมูลหลัง Cleansing ---")
display(df.head())

# STEP 8: Save to Excel
output_file = "data_cleansed.xlsx"
df.to_excel(output_file, index=False)
files.download(output_file)
```

สะกัด Hard Skills และ Soft Skills ออกจาก Job Description

```python
import pandas as pd
import re
from google.colab import files

# STEP 1: Upload file
print("กรุณาเลือกไฟล์ที่ต้องการอัปโหลด (CSV หรือ Excel):")
uploaded = files.upload()

# STEP 2: Load file
file_name = list(uploaded.keys())[0]

try:
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_name)
    elif file_name.endswith('.xlsx'):
        df = pd.read_excel(file_name)
    else:
        raise ValueError("⚠️ รองรับเฉพาะไฟล์ CSV และ Excel")
    print(f"✅ โหลดไฟล์ '{file_name}' เรียบร้อยแล้ว")
except Exception as e:
    print(f"เกิดข้อผิดพลาดในการโหลดไฟล์: {e}")
    raise

# STEP 3: Define skill finding function
def find_skill(description, skill):
    if pd.isna(description):
        return "No"
    pattern = r'\b' + re.escape(skill) + r'\b'
    return "Yes" if re.search(pattern, str(description), re.IGNORECASE) else "No"

# STEP 4: Define Hard + Soft Skills
hard_skills = [
    'python', 'r-studio', 'excel', 'sql', 'aws',
    'spark', 'tableau', 'power bi', 'machine learning', 'deep learning'
]

soft_skills = [
    'communication', 'teamwork', 'leadership', 'problem solving',
    'critical thinking', 'time management', 'adaptability',
    'creativity', 'collaboration', 'interpersonal skills'
]

all_skills = hard_skills + soft_skills

# STEP 5: Create skill columns (Yes/No)
df_clean = df.copy()
for skill in all_skills:
    col_name = skill.replace(' ', '_').replace('-', '_')
    df_clean[col_name] = df_clean['Job Description'].apply(lambda x: find_skill(x, skill))

# STEP 6: Export cleaned file to Excel
output_file_name = f"{file_name.split('.')[0]}_with_skills.xlsx"
df_clean.to_excel(output_file_name, index=False)

print("\n✅ ตรวจสอบ Hard & Soft Skills เสร็จสิ้น และสร้างไฟล์ใหม่เรียบร้อยแล้ว!")
print("📌 Preview (Job Title + Skills columns):")
print(df_clean[['Job Title'] + [s.replace(' ', '_').replace('-', '_') for s in all_skills]].head())

files.download(output_file_name)

```

Remote Extration

```python
import pandas as pd
import re
from google.colab import files

# STEP 1: Upload file
print("กรุณาเลือกไฟล์ที่ต้องการอัปโหลด (CSV หรือ Excel):")
uploaded = files.upload()

# STEP 2: Load file
file_name = list(uploaded.keys())[0]

if file_name.endswith('.csv'):
    df = pd.read_csv(file_name)
elif file_name.endswith('.xlsx'):
    df = pd.read_excel(file_name)
else:
    raise ValueError("⚠️ รองรับเฉพาะไฟล์ CSV และ Excel")

print(f"✅ โหลดไฟล์ '{file_name}' เรียบร้อยแล้ว")

# STEP 3: ฟังก์ชันตรวจสอบ Remote / Onsite
def get_job_remote(text):
    if pd.isna(text):
        return "Unknown"
    t = str(text).lower()
    if "remote" in t or "work form home" in t or "work from anywhere" in t :
        return "Remote"
    elif "onsite" in t or "on-site" in t or "office" in t:
        return "Onsite"
    elif "hybrid" in t :
        return "Hybrid"
    else:
        return "Unknown"

# STEP 4: สร้างคอลัมน์ Job Level
df["Job Remote"] = df.apply(
    lambda row: get_job_remote(str(row.get("Job Title", "")) + " " + str(row.get("Job Description", ""))),
    axis=1
)

# STEP 5: Export ไฟล์ใหม่
output_file_name = f"{file_name.split('.')[0]}_with_jobremote.xlsx"
df.to_excel(output_file_name, index=False)

print("\n✅ เพิ่มคอลัมน์ Job Level (Remote/Onsite) เรียบร้อยแล้ว!")
print(df[["Job Title", "Job Remote"]].head())

files.download(output_file_name)
```

Text Extration :  Level of Job Title, Experience in x years

```python
import pandas as pd
import re
from google.colab import files

# STEP 1: Upload file
print("กรุณาเลือกไฟล์ที่ต้องการอัปโหลด (CSV หรือ Excel):")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]

if file_name.endswith('.csv'):
    df = pd.read_csv(file_name)
elif file_name.endswith('.xlsx'):
    df = pd.read_excel(file_name)
else:
    raise ValueError("⚠️ รองรับเฉพาะไฟล์ CSV และ Excel")

print(f"✅ โหลดไฟล์ '{file_name}' เรียบร้อยแล้ว")

# STEP 2: Function หาระดับงานจาก Job Title
def extract_job_level(title):
    if pd.isna(title):
        return "Unknown"
    t = str(title).lower()
    if "intern" in t:
        return "Internship"
    elif "junior" in t or "associate" in t:
        return "Entry Level"
    elif "senior" in t or "lead" in t:
        return "Senior Level"
    elif "manager" in t or "director" in t:
        return "Management"
    elif "vp" in t or "vice president" in t or "chief" in t or "head" in t:
        return "Executive"
    else:
        return "Mid Level"

# STEP 3: Function หา Years of Experience
def extract_experience(text):
    if pd.isna(text):
        return "N/A"
    t = str(text).lower()
    match = re.search(r'(\d+)\+?\s*(?:years?|yrs?)\s+of\s+experience', t)
    if not match:
        match = re.search(r'(\d+)\s*(?:years?|yrs?)\s+experience', t)
    if match:
        return match.group(1)
    return "N/A"

# STEP 4: Apply functions
df["Job Level"] = df["Job Title"].apply(extract_job_level)
df["Experience (Years)"] = df.apply(
    lambda row: extract_experience(str(row.get("Job Title", "")) + " " + str(row.get("Job Description", ""))),
    axis=1
)

# STEP 5: Export
output_file_name = f"{file_name.split('.')[0]}_with_joblevel_experience.xlsx"
df.to_excel(output_file_name, index=False)

print("\n✅ Extract Job Level + Years of Experience เสร็จสิ้น!")
print(df[["Job Title", "Job Level", "Experience (Years)"]].head())

files.download(output_file_name)
```

Insight 1 : ตำแหน่งใดมีเงินเดือนสูงที่สุดในระดับเดียวกัน

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from textwrap import fill
from google.colab import files

# =========================
# STEP 1: Upload & Load
# =========================
print("กรุณาเลือกไฟล์ที่ต้องการอัปโหลด (Excel):")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)

# =========================
# STEP 2: Clean & Filter
# =========================
# มาตรฐานชื่อ level + แก้สะกดผิด Excecutive -> Executive
if "Job Level" in df.columns:
    df["Job Level"] = (
        df["Job Level"]
        .astype(str).str.strip()
        .replace({"Excecutive":"Executive"})
    )

# ตัดแถวที่ไม่มีค่า / เป็น 0
if "Salary Mid Point" not in df.columns:
    raise KeyError("ไม่พบคอลัมน์ 'Salary Mid Point' ในไฟล์ที่อัปโหลด")

df = df[df["Salary Mid Point"].notna()]
df = df[df["Salary Mid Point"] != 0]
df = df[df["Job Level"].notna()]

# highest salary per Job Level
highest_salary = df.loc[df.groupby("Job Level")["Salary Mid Point"].idxmax()].copy()

# ลำดับที่ต้องการ
level_order = ["Entry Level", "Senior Level", "Mid Level", "Management", "Executive"]
highest_salary = highest_salary[highest_salary["Job Level"].isin(level_order)]
highest_salary["Job Level"] = pd.Categorical(highest_salary["Job Level"],
                                             categories=level_order, ordered=True)
highest_salary = highest_salary.sort_values("Job Level")

# =========================
# STEP 3: Chart (Pastel + LEFT labels, FULL titles)
# =========================
fig, ax = plt.subplots(figsize=(12, 6))

vals = highest_salary["Salary Mid Point"].astype(float).values
xmax = vals.max() * 1.12
ax.set_xlim(0, xmax)

# โทนสีละมุน (PuBuGn)
cmap = plt.cm.PuBuGn
norm = plt.Normalize(vals.min(), vals.max())
colors = cmap(norm(vals))

bars = ax.barh(
    highest_salary["Job Level"], vals,
    color=colors, edgecolor="#dfe7ef", linewidth=0.8, alpha=0.9
)

# ทำให้ Entry Level เข้มขึ้นเล็กน้อย
for bar, lvl in zip(bars, highest_salary["Job Level"].astype(str)):
    if lvl == "Entry Level":
        bar.set_facecolor(plt.cm.PuBuGn(0.75))
        bar.set_alpha(0.95)

# เส้นกริดบาง ๆ
ax.xaxis.grid(True, linestyle=(0,(2,4)), linewidth=0.7, color="#b9c7d8", alpha=0.5)
ax.set_axisbelow(True)
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)
ax.spines["left"].set_color("#c9d3df")
ax.spines["bottom"].set_color("#c9d3df")
ax.spines["left"].set_linewidth(0.8)
ax.spines["bottom"].set_linewidth(0.8)

# ฟอร์แมตแกน X
ax.xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
ax.set_xlabel("Salary Mid Point", color="#334155")
ax.set_ylabel("Job Level", color="#334155")
ax.set_title("Highest Salary by Job Level (using Salary Mid Point)", color="#1f2937", pad=14)

# =========================
# Labels: ชิดซ้ายของกราฟ, ใช้ชื่อเต็ม + จำนวนเงินในวงเล็บ
# =========================
LEFT_PAD = xmax * 0.01  # ระยะจากขอบซ้ายของกราฟ
WRAP_WIDTH = 62         # ตัดบรรทัดอัตโนมัติเมื่อชื่อยาวมาก (ยังคงชื่อเต็ม)

for bar, title, val in zip(bars, highest_salary["Job Title"], vals):
    full_title = str(title)
    text_str = f"{full_title}  ({val:,.0f})"  # จำนวนเงินตามตัว ใส่วงเล็บ
    # wrap ชื่อยาวให้ขึ้นบรรทัดใหม่ แต่ไม่ตัดคำทิ้ง
    text_str = fill(text_str, width=WRAP_WIDTH)
    y_text = bar.get_y() + bar.get_height()/2
    ax.text(
        LEFT_PAD, y_text, text_str,
        ha="left", va="center", color="#111827", fontsize=10
    )

plt.tight_layout()
plt.show()

# =========================
# STEP 4: Export
# =========================
out_name = "highest_salary_by_joblevel.xlsx"
highest_salary.to_excel(out_name, index=False)
files.download(out_name)
```

```python
# =========================
# Colab: Size vs Salary Mid Point
# =========================
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from google.colab import files

plt.rcParams.update({"figure.dpi": 130, "font.size": 10})

# === Upload file ===
print("กรุณาอัปโหลดไฟล์ Excel (เช่น Data ETL Final_state and rating.xlsx)")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# === Load ===
df = pd.read_excel(file_name)
df.columns = df.columns.str.strip()

# === ตั้งชื่อคอลัมน์หลัก ===
COL_SAL = "Salary Mid Point"
COL_SIZE = "Size"

# === ตรวจคอลัมน์เงินเดือน ===
if COL_SAL not in df.columns:
    raise KeyError("ไม่พบคอลัมน์ 'Salary Mid Point' ในไฟล์")

# แปลง Salary Mid Point เป็นตัวเลข และกรองค่า 0/NaN
df[COL_SAL] = pd.to_numeric(df[COL_SAL], errors="coerce")
df = df[df[COL_SAL].notna() & (df[COL_SAL] > 0)]

# === Normalize 'Size' → เป็นช่วงมาตรฐาน ===
def normalize_size(x: str) -> str:
    if pd.isna(x):
        return "Unknown"
    s = str(x).lower().strip()
    s = s.replace("employees", "").replace("employee", "").strip()
    s = s.replace(" to ", "-").replace("–", "-").replace("—", "-")
    s = re.sub(r"\s+", " ", s)

    # "10000+"
    m_plus = re.match(r"^(\d+)\+\s*$", s)
    if m_plus:
        lo = int(m_plus.group(1))
        return f"{lo}+"
    # "x-y"
    m_rng = re.match(r"^(\d+)\s*-\s*(\d+)$", s)
    if m_rng:
        lo, hi = int(m_rng.group(1)), int(m_rng.group(2))
        return f"{lo}-{hi}"
    # single number
    m_single = re.match(r"^\d+$", s)
    if m_single:
        n = int(m_single.group(0))
        return f"{n}-{n}"
    return "Unknown"

if COL_SIZE in df.columns:
    df["Size_norm"] = df[COL_SIZE].apply(normalize_size)
else:
    df["Size_norm"] = "Unknown"

# ลำดับช่วงที่ต้องการแสดงบนกราฟ (ปรับได้)
size_order = [
    "1-50", "51-200", "201-500", "501-1000",
    "1001-5000", "5001-10000", "10000+"
]
# ค่าที่ไม่อยู่ในลิสต์ → Unknown
df.loc[~df["Size_norm"].isin(size_order), "Size_norm"] = "Unknown"
df["Size_norm"] = pd.Categorical(df["Size_norm"], categories=size_order + ["Unknown"], ordered=True)

# === ตารางสถิติ ===
size_stats = (df
              .groupby("Size_norm", dropna=False)[COL_SAL]
              .agg(count="count", mean="mean", median="median", std="std")
              .reset_index()
             )

print("สถิติ Salary Mid Point แยกตามขนาดบริษัท (Size):")
display(size_stats)

# === เตรียมลำดับสำหรับกราฟ ===
ordered = [c for c in size_order if c in df["Size_norm"].unique()]
if "Unknown" in df["Size_norm"].unique():
    ordered += ["Unknown"]

# === กราฟ Boxplot: การกระจายเงินเดือนในแต่ละ Size ===
fig, ax = plt.subplots(figsize=(9, 5))
data_for_box = [df.loc[df["Size_norm"] == cat, COL_SAL].values for cat in ordered]
ax.boxplot(data_for_box, labels=ordered, patch_artist=True)
ax.set_title("Distribution of Salary Mid Point by Company Size")
ax.set_ylabel("Salary Mid Point")
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
ax.grid(True, axis="y", linestyle="--", alpha=0.35)
plt.tight_layout()
plt.show()

# === กราฟ Bar: ค่าเฉลี่ยเงินเดือนในแต่ละ Size ===
fig, ax = plt.subplots(figsize=(9, 5))
means = size_stats.set_index("Size_norm").loc[ordered, "mean"]
bars = ax.bar(ordered, means)
ax.set_title("Average Salary Mid Point by Company Size")
ax.set_ylabel("Average Salary Mid Point")
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
ax.grid(True, axis="y", linestyle="--", alpha=0.35)

# ใส่ตัวเลขบนแท่ง
for x, v in enumerate(means):
    ax.text(x, v, f"{v:,.0f}", ha="center", va="bottom")

plt.tight_layout()
plt.show()

# === Export สรุปเป็น Excel ===
out_xlsx = "size_vs_salary_summary.xlsx"
with pd.ExcelWriter(out_xlsx, engine="xlsxwriter") as writer:
    size_stats.to_excel(writer, sheet_name="Size_vs_Salary", index=False)

files.download(out_xlsx)
```

```python
# ===============================================================
# HR Insights — Full Colab Version (Upload + Flexible Loader + Anti-Overlap Charts + RAI)
# ===============================================================
# What you get:
# 1) Cleaned aggregates & CSVs:
#    - state_summary.csv, joblevel_pay_compression.csv, company_size_pay.csv
#    - ownership_pay.csv, industry_pay.csv, sector_pay.csv
#    - state_colid_salary_rpp.csv, skill_demand.csv, skill_salary_premium.csv
#    - company_rpp.csv
# 2) Charts (PNG; English labels, font size 14; anti-overlap):
#    - Headcount by State, COLID by State, Salary by State, Dual COLID+Salary
#    - Scatter COLID vs Salary, RPP Histogram
#    - Job Level Median Salary, Pay Compression
#    - Company Size / Ownership / Industry / Sector vs Salary
#    - Skill Demand, Skill Salary Premium, Company RPP
# 3) Role Attractiveness Index (RAI):
#    - role_attractiveness_index_all.csv
#    - role_attractiveness_top10.csv, role_attractiveness_bottom10.csv
#    - role_attractiveness_onepager.md
# 4) Auto-download ZIP of outputs in Colab
# ---------------------------------------------------------------

import pandas as pd
import numpy as np
import re, textwrap, os, io, zipfile, shutil
from pathlib import Path
import matplotlib
import matplotlib.pyplot as plt

# ============== Config (English graph labels, font size 14) ==============
matplotlib.rcParams.update({
    "font.size": 14,
    "axes.titlesize": 14,
    "axes.labelsize": 14,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
    "legend.fontsize": 14,
    "axes.unicode_minus": False,
})

OUT_DIR = Path("./hr_outputs")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# =================== (NEW) Colab Upload Block ===================
# รันใน Colab: เด้งกล่องให้อัปโหลด Excel (และไฟล์อื่นที่อยากอัปได้)
try:
    from google.colab import files
    print("▶ Please upload your Excel file now (Rev03/Rev02, .xlsx/.xls). You can upload multiple files:")
    _uploaded = files.upload()  # อัปโหลดแล้วจะมีไฟล์ไปอยู่ใน working dir
except Exception:
    _uploaded = {}

# ========================= Helpers =========================
def _norm(s: str) -> str:
    return re.sub(r'[^a-z0-9ก-๙]+', '', str(s).strip().lower())

def pick_col(df, aliases):
    normmap = {_norm(c): c for c in df.columns}
    for a in aliases:
        k = _norm(a)
        if k in normmap:
            return normmap[k]
    for a in sorted(aliases, key=len, reverse=True):
        ak = _norm(a)
        for k, v in normmap.items():
            if ak and ak in k:
                return v
    return None

def to_number(series):
    # supports raw numbers and suffix: k/m/b
    s = series.astype(str).str.replace(r"[^\d\.\-kKmMbB]", "", regex=True)
    def parse_unit(x):
        x = str(x).strip()
        if x == "" or x.lower() == "nan": return np.nan
        mult = 1.0
        last = x[-1:].lower()
        if last == 'k': mult, x = 1e3, x[:-1]
        elif last == 'm': mult, x = 1e6, x[:-1]
        elif last == 'b': mult, x = 1e9, x[:-1]
        try:
            return float(x) * mult
        except:
            try:
                return float(x.replace(",","")) * mult
            except:
                return np.nan
    return s.apply(parse_unit)

def parse_company_size(value):
    if pd.isna(value): return np.nan
    s = str(value).strip().lower()
    if "small" in s: return "Small"
    if "medium" in s or "mid" in s: return "Medium"
    if "large" in s or "enterprise" in s: return "Large"
    nums = re.findall(r"\d+", s)
    if not nums: return str(value)
    nums = list(map(int, nums))
    if "+" in s and nums: head = nums[0]
    elif len(nums) >= 2:  head = (nums[0] + nums[1]) / 2.0
    else:                 head = nums[0]
    if head <= 50:   return "1-50"
    if head <= 200:  return "51-200"
    if head <= 500:  return "201-500"
    if head <= 1000: return "501-1k"
    if head <= 5000: return "1k-5k"
    return "5k+"

def skill_bool(series):
    s = series.fillna(0)
    if s.dtype == bool:
        return s
    def to_bool(x):
        try:
            if float(x) > 0: return True
        except:
            pass
        ss = str(x).strip().lower()
        return ss in {"y","yes","true","t","required","need","must","1"}
    return s.apply(to_bool)

def wrap_tick_labels(ax, axis="x", width=12, rotation=None):
    if axis == "x":
        labels = ax.get_xticklabels()
        for lab in labels:
            lab.set_text(textwrap.fill(lab.get_text(), width=width))
        ax.set_xticklabels(labels, rotation=(rotation if rotation is not None else 0),
                           ha="right" if rotation else "center")
    else:
        labels = ax.get_yticklabels()
        for lab in labels:
            lab.set_text(textwrap.fill(lab.get_text(), width=width))
        ax.set_yticklabels(labels)

def set_dynamic_figsize(n, orientation="v", base_w=8, base_h=6, per_bar=0.45, max_w=26, max_h=18):
    if orientation == "v":
        w = min(max(base_w, n * per_bar), max_w); h = base_h
    else:
        w = base_w + 2; h = min(max(base_h, n * per_bar), max_h)
    plt.gcf().set_size_inches(w, h)

def savefig(fname):
    out = OUT_DIR/fname
    plt.tight_layout()
    plt.savefig(out, bbox_inches="tight", dpi=144)
    plt.close()
    return out.as_posix()

def plot_series_cat(series, title, xlabel, ylabel, fname, prefer_horizontal_over_n=12, wrap_width=12):
    series = series.copy()
    n = len(series)
    if n == 0: return None
    use_h = n > prefer_horizontal_over_n
    plt.figure()
    set_dynamic_figsize(n, "h" if use_h else "v")
    if use_h:
        plt.barh(series.index, series.values)
        ax = plt.gca()
        ax.invert_yaxis()
        ax.set_title(title); ax.set_xlabel(xlabel); ax.set_ylabel(ylabel)
        wrap_tick_labels(ax, axis="y", width=wrap_width)
        plt.margins(x=0.02, y=0.02)
    else:
        series.plot(kind="bar")
        ax = plt.gca()
        ax.set_title(title); ax.set_xlabel(ylabel); ax.set_ylabel(xlabel)
        wrap_tick_labels(ax, axis="x", width=wrap_width, rotation=45)
        plt.margins(x=0.01, y=0.05)
    return savefig(fname)

def resolve_excel_path():
    # Try expected names first
    candidates_exact = [
        "Data ETL Final_state and rating_Rev03.xlsx",
        "Data ETL Final_state and rating_Rev02.xlsx",
        "Data ETL Final_state and rating.xlsx",
    ]
    search_dirs = [Path("/content"), Path("/mnt/data"), Path(".")]
    for d in search_dirs:
        for name in candidates_exact:
            p = d / name
            if p.exists(): return p
    # Fuzzy search
    found = []
    tokens = ["final", "state", "rating", "rev03", "rev02", "etl"]
    for d in search_dirs:
        if not d.exists(): continue
        for pat in ("*.xlsx", "*.xls"):
            for p in d.glob(pat):
                name = p.name.lower()
                score = sum(t in name for t in tokens)
                found.append((score, -p.stat().st_size if p.exists() else 0, p))
    if found:
        found.sort(reverse=True)
        return found[0][2]
    # Last chance: if running in Colab, prompt upload now
    try:
        from google.colab import files
        print("Excel not found by search. Please upload your Excel (.xlsx/.xls) now:")
        uploaded = files.upload()
        for k in uploaded.keys():
            if k.lower().endswith((".xlsx", ".xls")):
                return Path(k)
    except Exception:
        pass
    return None

def safe_download(path_like):
    # Best effort download in Colab; no-op elsewhere
    try:
        from google.colab import files
        files.download(str(path_like))
    except Exception:
        pass

# ====================== Load Excel (auto-best sheet) ======================
xlsx_path = resolve_excel_path()
assert xlsx_path is not None and xlsx_path.exists(), "Excel file not found. Please upload the Excel."
print("Using Excel:", xlsx_path)

xls = pd.ExcelFile(xlsx_path)
alias_bank = [
    "Company Name","Location","State","Desc","COLID","Size","Type of ownership","Industry","Sector",
    "Revenue","Job Title","Job Level","Job Category","Rating",
    "Salary Min","Salary Max","Salary Mid Point",
    "python","r_studio","excel","sql","aws","spark","tableau","power_bi","machine_learning",
    "deep_learning","communication","teamwork","leadership","problem_solving","critical_thinking",
    "time_management","adaptability","creativity","collaboration","interpersonal_skills"
]
alias_norm = set(_norm(a) for a in alias_bank)

best_sheet, best_score = None, -1
for sh in xls.sheet_names:
    df_try = pd.read_excel(xls, sh)
    score = len({_norm(c) for c in df_try.columns} & alias_norm)
    if score > best_score:
        best_score, best_sheet = score, sh

df = pd.read_excel(xls, best_sheet)
print("Sheet used:", best_sheet)

# ====================== Column mapping ======================
col_company   = pick_col(df, ["Company Name","company","employer"])
col_location  = pick_col(df, ["Location"])
col_state     = pick_col(df, ["State"])
col_desc      = pick_col(df, ["Desc","State Name","State Full"])
col_colid     = pick_col(df, ["COLID","Cost of living index","Composite Index","COLI"])
col_size      = pick_col(df, ["Size","Company Size","Employer Size"])
col_owner     = pick_col(df, ["Type of ownership","Ownership","Type"])
col_industry  = pick_col(df, ["Industry"])
col_sector    = pick_col(df, ["Sector"])
col_revenue   = pick_col(df, ["Revenue"])
col_title     = pick_col(df, ["Job Title","Title","Position"])
col_level     = pick_col(df, ["Job Level","Level","Grade","Band"])
col_category  = pick_col(df, ["Job Category","Category","Function"])
col_rating    = pick_col(df, ["Rating","Employee Rating","Score"])
col_salmin    = pick_col(df, ["Salary Min","Min Salary","Lower Salary"])
col_salmax    = pick_col(df, ["Salary Max","Max Salary","Upper Salary"])
col_salmid    = pick_col(df, ["Salary Mid Point","Mid Salary","Salary Mid"])

skills_list = [
    "python","r_studio","excel","sql","aws","spark","tableau","power_bi","machine_learning",
    "deep_learning","communication","teamwork","leadership","problem_solving","critical_thinking",
    "time_management","adaptability","creativity","collaboration","interpersonal_skills"
]
present_skills = []
for sk in skills_list:
    col = pick_col(df, [sk])
    if col:
        present_skills.append((sk, col))

# ====================== Cleaning & Feature build ======================
if col_rating: df[col_rating] = pd.to_numeric(df[col_rating], errors="coerce")
for c in [col_colid, col_salmin, col_salmax, col_salmid]:
    if c: df[c] = to_number(df[c])

# Salary mid fallback
if col_salmid is None and (col_salmin or col_salmax):
    cols = [c for c in [col_salmin, col_salmax] if c]
    df["salary_mid_calc"] = df[cols].mean(axis=1, skipna=True)
    col_salmid = "salary_mid_calc"

# Size bucket
df["size_bucket"] = df[col_size].apply(parse_company_size) if col_size else np.nan

# Real Pay Proxy
df["RPP"] = np.where(
    col_salmid and col_colid and df[col_colid].notna() & df[col_salmid].notna() & (df[col_colid] > 0),
    df[col_salmid] / (df[col_colid] / 100.0),
    np.nan
)

# ====================== Aggregates (CSV) ======================
csv_paths = {}

# State summary
if col_state:
    state_summary = df.groupby(col_state).agg(
        headcount=(col_state,"size"),
        avg_rating=(col_rating,"mean") if col_rating else (col_state,"size"),
        avg_salary_mid=(col_salmid,"mean") if col_salmid else (col_state,"size"),
        avg_COLID=(col_colid,"mean") if col_colid else (col_state,"size"),
        avg_RPP=("RPP","mean") if "RPP" in df.columns else (col_state,"size")
    ).reset_index().rename(columns={col_state:"state"}).sort_values("headcount", ascending=False)
    p = OUT_DIR/"state_summary.csv"; state_summary.to_csv(p, index=False); csv_paths["state_summary"]=p.as_posix()
else:
    state_summary = pd.DataFrame()

# Job level median + compression
if col_level and col_salmid:
    jl = df.groupby(col_level)[col_salmid].median().dropna().sort_values(ascending=False).rename("median_salary_mid")
    jl = jl.reset_index().rename(columns={col_level:"job_level"})
    jl["next_level_median"] = jl["median_salary_mid"].shift(-1)
    jl["compression_pct_to_next"] = (jl["next_level_median"]/jl["median_salary_mid"] - 1.0)*100.0
    p = OUT_DIR/"joblevel_pay_compression.csv"; jl.to_csv(p, index=False); csv_paths["joblevel_pay_compression"]=p.as_posix()

def group_pay(colname, outname):
    if colname and col_salmid:
        gp = df.groupby(colname)[col_salmid].agg(["count","median","mean"]).sort_values("median", ascending=False).reset_index()
        gp.columns = [colname, "count", "median_salary_mid", "mean_salary_mid"]
        p = OUT_DIR/outname; gp.to_csv(p, index=False); csv_paths[outname]=p.as_posix()
        return gp
    return pd.DataFrame()

ownership_pay = group_pay(col_owner,   "ownership_pay.csv")
industry_pay  = group_pay(col_industry,"industry_pay.csv")
sector_pay    = group_pay(col_sector,  "sector_pay.csv")
size_pay      = group_pay("size_bucket","company_size_pay.csv")

# State COLID vs Salary vs RPP
if col_colid and col_salmid and col_state:
    state_pair = df.groupby(col_state).agg(
        avg_colid=(col_colid,"mean"),
        avg_salary=(col_salmid,"mean"),
        avg_rpp=("RPP","mean")
    ).reset_index().rename(columns={col_state:"state"})
    p = OUT_DIR/"state_colid_salary_rpp.csv"; state_pair.to_csv(p, index=False); csv_paths["state_colid_salary_rpp"]=p.as_posix()
else:
    state_pair = pd.DataFrame()

# Skills: demand & premium
def skill_demand_and_premium(df, present_skills, level_col, sal_col):
    if not present_skills or sal_col is None: return pd.DataFrame(), pd.DataFrame()
    dem_rows = []
    for sk, col in present_skills:
        req = skill_bool(df[col]); dem_rows.append((sk, int(req.sum())))
    demand_df = pd.DataFrame(dem_rows, columns=["skill","count_required"]).sort_values("count_required", ascending=False)

    if level_col:
        prem_rows = []
        for sk, col in present_skills:
            req_mask = skill_bool(df[col])
            tmp = df[[level_col, sal_col]].copy(); tmp["req"]=req_mask
            tmp = tmp.dropna(subset=[sal_col])
            diffs, weights = [], []
            for lvl, g in tmp.groupby(level_col):
                g_req = g[g["req"]==True][sal_col]; g_not = g[g["req"]==False][sal_col]
                if len(g_req) >= 5 and len(g_not) >= 5:
                    diffs.append(g_req.mean()-g_not.mean()); weights.append(len(g))
            if diffs: wavg = float(np.average(diffs, weights=weights))
            else:     wavg = np.nan
            prem_rows.append((sk, wavg, int(np.sum(weights)) if weights else 0))
        premium_df = pd.DataFrame(prem_rows, columns=["skill","weighted_diff_vs_no_skill","obs_used"]) \
                        .sort_values("weighted_diff_vs_no_skill", ascending=False)
    else:
        premium_df = pd.DataFrame()

    return demand_df, premium_df

skill_demand_df, skill_premium_df = skill_demand_and_premium(df, present_skills, col_level, col_salmid)
if not skill_demand_df.empty:
    p = OUT_DIR/"skill_demand.csv"; skill_demand_df.to_csv(p, index=False); csv_paths["skill_demand"]=p.as_posix()
if not skill_premium_df.empty:
    p = OUT_DIR/"skill_salary_premium.csv"; skill_premium_df.to_csv(p, index=False); csv_paths["skill_salary_premium"]=p.as_posix()

# Company RPP
if col_company and col_salmid and col_colid:
    company_rpp = df.groupby(col_company).agg(
        headcount=(col_company,"size"),
        avg_rpp=("RPP","mean"),
        median_salary=(col_salmid,"median")
    ).reset_index().rename(columns={col_company:"company"}).sort_values("avg_rpp", ascending=False)
    p = OUT_DIR/"company_rpp.csv"; company_rpp.to_csv(p, index=False); csv_paths["company_rpp"]=p.as_posix()
else:
    company_rpp = pd.DataFrame()

# ====================== Charts (anti-overlap) ======================
charts = []

# A) State headcount
if col_state:
    s = df[col_state].value_counts(dropna=False).sort_values(ascending=False).head(20)
    charts.append(plot_series_cat(s, "Headcount by State (Top 20)",
                                  xlabel="Records", ylabel="State",
                                  fname="chart_state_headcount_top20.png",
                                  prefer_horizontal_over_n=10, wrap_width=10))

# B) Average COLID by State (Top 20)
if col_state and col_colid:
    s = df.groupby(col_state)[col_colid].mean().dropna().sort_values(ascending=False).head(20)
    charts.append(plot_series_cat(s, "Average Cost of Living Index by State (Top 20)",
                                  xlabel="Average COLID", ylabel="State",
                                  fname="chart_state_avg_COLID_top20.png",
                                  prefer_horizontal_over_n=10, wrap_width=10))

# C) Average Salary (Mid) by State (Top 20)
if col_state and col_salmid:
    s = df.groupby(col_state)[col_salmid].mean().dropna().sort_values(ascending=False).head(20)
    charts.append(plot_series_cat(s, "Average Salary (Mid Point) by State (Top 20)",
                                  xlabel="Average Salary (Mid)", ylabel="State",
                                  fname="chart_state_avg_salary_top20.png",
                                  prefer_horizontal_over_n=10, wrap_width=10))

# D) Dual Axis: COLID + Salary (Top 20 by COLID)
if col_state and col_salmid and col_colid:
    pair = df.groupby(col_state).agg(avg_colid=(col_colid,"mean"),
                                     avg_salary=(col_salmid,"mean")).dropna()
    top = pair.sort_values("avg_colid", ascending=False).head(20)
    plt.figure(); set_dynamic_figsize(len(top), "v")
    ax1 = plt.gca()
    ax1.bar(top.index, top["avg_colid"].values)
    ax1.set_title("Average COLID + Average Salary by State (Top 20 by COLID)")
    ax1.set_xlabel("State"); ax1.set_ylabel("Average COLID")
    wrap_tick_labels(ax1, axis="x", width=10, rotation=45)
    ax2 = ax1.twinx(); ax2.plot(top.index, top["avg_salary"].values, marker="o"); ax2.set_ylabel("Average Salary (Mid)")
    charts.append(savefig("chart_state_dual_COLID_salary_top20.png"))

# E) Scatter: Average COLID vs Average Salary (by State)
if col_state and col_salmid and col_colid:
    pair = df.groupby(col_state).agg(avg_colid=(col_colid,"mean"),
                                     avg_salary=(col_salmid,"mean")).dropna()
    plt.figure(); set_dynamic_figsize(max(10, len(pair)//2), "v")
    plt.scatter(pair["avg_colid"], pair["avg_salary"])
    plt.title("Average COLID vs Average Salary (by State)")
    plt.xlabel("Average COLID"); plt.ylabel("Average Salary (Mid)")
    if len(pair) >= 3:
        x = pair["avg_colid"].values; y = pair["avg_salary"].values
        b1, b0 = np.polyfit(x, y, 1)
        xx = np.linspace(x.min(), x.max(), 100); yy = b1*xx + b0
        plt.plot(xx, yy)
    charts.append(savefig("chart_scatter_COLID_vs_salary_state.png"))

# F) Histogram: RPP
if "RPP" in df.columns and df["RPP"].notna().sum() >= 10:
    plt.figure(); set_dynamic_figsize(10, "v")
    df["RPP"].dropna().plot(kind="hist", bins=30)
    plt.title("Real Pay Proxy Distribution (Salary Mid / (COLID/100))")
    plt.xlabel("RPP"); plt.ylabel("Frequency")
    charts.append(savefig("chart_hist_RPP.png"))

# G) Job Level vs Median Salary
if col_level and col_salmid:
    s = df.groupby(col_level)[col_salmid].median().dropna().sort_values(ascending=False)
    charts.append(plot_series_cat(s, "Job Level vs Median Salary (Mid Point)",
                                  xlabel="Median Salary (Mid)", ylabel="Job Level",
                                  fname="chart_joblevel_median_salary.png",
                                  prefer_horizontal_over_n=8, wrap_width=14))

# H) Pay Compression to Next Level
if col_level and col_salmid:
    jl2 = df.groupby(col_level)[col_salmid].median().dropna().sort_values(ascending=False)
    jl2 = jl2.reset_index().rename(columns={col_level:"job_level", col_salmid:"median_salary_mid"})
    jl2["next_level_median"] = jl2["median_salary_mid"].shift(-1)
    jl2["compression_pct_to_next"] = (jl2["next_level_median"]/jl2["median_salary_mid"] - 1.0) * 100.0
    ser = jl2.set_index("job_level")["compression_pct_to_next"]
    charts.append(plot_series_cat(ser, "Pay Compression to Next Level (%)",
                                  xlabel="Compression % to Next Level", ylabel="Job Level",
                                  fname="chart_joblevel_compression.png",
                                  prefer_horizontal_over_n=8, wrap_width=14))

# I) Company Size vs Average Salary
if "size_bucket" in df.columns and col_salmid:
    s = df.groupby("size_bucket")[col_salmid].mean().dropna().sort_values(ascending=False)
    charts.append(plot_series_cat(s, "Company Size vs Average Salary (Mid)",
                                  xlabel="Average Salary (Mid)", ylabel="Company Size Bucket",
                                  fname="chart_companysize_avg_salary.png",
                                  prefer_horizontal_over_n=8, wrap_width=10))

# J) Ownership Type vs Median Salary (Top 12)
if col_owner and col_salmid:
    s = df.groupby(col_owner)[col_salmid].median().dropna().sort_values(ascending=False).head(12)
    charts.append(plot_series_cat(s, "Ownership Type vs Median Salary (Top 12)",
                                  xlabel="Median Salary (Mid)", ylabel="Ownership Type",
                                  fname="chart_ownership_median_salary_top12.png",
                                  prefer_horizontal_over_n=8, wrap_width=14))

# K) Industry vs Median Salary (Top 15)
if col_industry and col_salmid:
    s = df.groupby(col_industry)[col_salmid].median().dropna().sort_values(ascending=False).head(15)
    charts.append(plot_series_cat(s, "Industry vs Median Salary (Top 15)",
                                  xlabel="Median Salary (Mid)", ylabel="Industry",
                                  fname="chart_industry_median_salary_top15.png",
                                  prefer_horizontal_over_n=8, wrap_width=16))

# L) Sector vs Median Salary (Top 12)
if col_sector and col_salmid:
    s = df.groupby(col_sector)[col_salmid].median().dropna().sort_values(ascending=False).head(12)
    charts.append(plot_series_cat(s, "Sector vs Median Salary (Top 12)",
                                  xlabel="Median Salary (Mid)", ylabel="Sector",
                                  fname="chart_sector_median_salary_top12.png",
                                  prefer_horizontal_over_n=8, wrap_width=16))

# M) Skill Demand (Top 15)
if present_skills:
    dem = []
    for sk, col in present_skills:
        req = skill_bool(df[col]); dem.append((sk, int(req.sum())))
    demand_df = pd.DataFrame(dem, columns=["skill","count_required"]).sort_values("count_required", ascending=False).head(15)
    if not demand_df.empty:
        s = demand_df.set_index("skill")["count_required"]
        charts.append(plot_series_cat(s, "Skill Demand (Top 15 by Required Count)",
                                      xlabel="Count Required", ylabel="Skill",
                                      fname="chart_skill_demand_top15.png",
                                      prefer_horizontal_over_n=8, wrap_width=14))

# N) Skill Salary Premium (Top 15; within-level weighted)
if present_skills and col_salmid:
    prem_rows = []
    for sk, col in present_skills:
        req_mask = skill_bool(df[col])
        tmp = df[[col_level, col_salmid]].copy() if col_level else df[[col_salmid]].copy()
        tmp["req"] = req_mask; tmp = tmp.dropna(subset=[col_salmid])
        if tmp.empty:
            prem_rows.append((sk, np.nan, 0)); continue
        if col_level:
            diffs, weights = [], []
            for lvl, g in tmp.groupby(col_level):
                g_req = g[g["req"]==True][col_salmid]; g_not = g[g["req"]==False][col_salmid]
                if len(g_req) >= 5 and len(g_not) >= 5:
                    diffs.append(g_req.mean()-g_not.mean()); weights.append(len(g))
            wavg = float(np.average(diffs, weights=weights)) if diffs else np.nan
            prem_rows.append((sk, wavg, int(np.sum(weights)) if weights else 0))
        else:
            prem_rows.append((sk, np.nan, 0))
    prem_df = pd.DataFrame(prem_rows, columns=["skill","weighted_diff_vs_no_skill","obs_used"]) \
                .sort_values("weighted_diff_vs_no_skill", ascending=False).head(15)
    if not prem_df.empty:
        s = prem_df.set_index("skill")["weighted_diff_vs_no_skill"]
        charts.append(plot_series_cat(s, "Skill Salary Premium (Top 15)",
                                      xlabel="Weighted Diff vs No Skill", ylabel="Skill",
                                      fname="chart_skill_salary_premium_top15.png",
                                      prefer_horizontal_over_n=8, wrap_width=14))

# O) Company RPP (Top 15)
if col_company and "RPP" in df.columns and df["RPP"].notna().any():
    comp = df.groupby(col_company).agg(avg_rpp=("RPP","mean"), headcount=(col_company,"size")).dropna().sort_values("avg_rpp", ascending=False).head(15)
    ser = comp["avg_rpp"]; ser.index = comp.index.astype(str)
    charts.append(plot_series_cat(ser, "Company Real Pay Proxy (Top 15 by Avg RPP)",
                                  xlabel="Avg RPP", ylabel="Company",
                                  fname="chart_company_rpp_top15.png",
                                  prefer_horizontal_over_n=6, wrap_width=18))

# ====================== Role Attractiveness Index (RAI) ======================
# Percentiles within State×Job Level
if col_state and col_level:
    grp = df.groupby([col_state, col_level], dropna=False)
    df["pct_RPP_state_level"] = grp["RPP"].transform(lambda s: s.rank(pct=True)*100)
    df["pct_pay_state_level"] = grp[col_salmid].transform(lambda s: s.rank(pct=True)*100)
else:
    df["pct_RPP_state_level"] = np.nan
    df["pct_pay_state_level"] = np.nan

# Compression penalty
if col_level and col_salmid:
    jl_med = df.groupby(col_level)[col_salmid].median().dropna().sort_values(ascending=False)
    tab = jl_med.reset_index(); tab.columns = ["job_level","median_salary"]
    tab["next_median"] = tab["median_salary"].shift(-1)
    tab["compression_pct_to_next"] = (tab["next_median"]/tab["median_salary"] - 1.0)*100.0
    comp_map = tab.set_index("job_level")["compression_pct_to_next"].to_dict()
    df["compression_pct_to_next"] = df[col_level].map(comp_map)
else:
    df["compression_pct_to_next"] = np.nan

def compression_penalty(x):
    if pd.isna(x): return 30.0
    if x >= 12:  return 0.0
    if x >= 8:   return 30.0
    if x >= 0:   return 60.0
    return 100.0

df["penalty_compression"] = df["compression_pct_to_next"].apply(compression_penalty)

# Premium skill coverage (share of required premium skills)
skill_cols = [c for c in df.columns if str(c).strip().lower() in set(skills_list)]
premium_skills = set()
for c in skill_cols:
    req = skill_bool(df[c])
    with_s = df.loc[req, col_salmid]; without = df.loc[~req, col_salmid]
    if with_s.notna().sum() >= 10 and without.notna().sum() >= 10:
        diff = with_s.mean() - without.mean()
        if diff > 0: premium_skills.add(str(c).strip().lower())

def share_premium_skills_row(row):
    if not premium_skills: return 0.0
    have, tot = 0, 0
    for c in skill_cols:
        tot += 1
        if str(c).strip().lower() in premium_skills and skill_bool(df[c]).iloc[row.name]:
            have += 1
    return (have/tot*100.0) if tot>0 else 0.0

df["pct_premium_skill_coverage"] = df.apply(share_premium_skills_row, axis=1)

# Rating normalized
df["rating_norm"] = (df[col_rating]/5.0)*100.0 if col_rating else np.nan

# RAI formula (0..100)
w = dict(rpp=0.35, pay=0.20, skill=0.20, rating=0.15, penalty=-0.10)
for c in ["pct_RPP_state_level","pct_pay_state_level","pct_premium_skill_coverage","rating_norm","penalty_compression"]:
    m = df[c].mean(skipna=True); df[c] = df[c].fillna(m)

df["RAI"] = (
    w["rpp"]    * df["pct_RPP_state_level"] +
    w["pay"]    * df["pct_pay_state_level"] +
    w["skill"]  * df["pct_premium_skill_coverage"] +
    w["rating"] * df["rating_norm"] +
    w["penalty"]* df["penalty_compression"]
).clip(0, 100)

grp_cols = [c for c in [col_title, col_level, col_state] if c]
summary = df.groupby(grp_cols).agg(
    n=("RAI","size"),
    RAI_mean=("RAI","mean"),
    RAI_median=("RAI","median"),
    RPP_median=("RPP","median"),
    Pay_median=(col_salmid,"median"),
    COLID_median=(col_colid,"median"),
    Rating_median=(col_rating,"median") if col_rating else ("RAI","median"),
    pct_RPP_p50=("pct_RPP_state_level","median"),
    pct_Pay_p50=("pct_pay_state_level","median"),
    pct_PremiumSkill_p50=("pct_premium_skill_coverage","median"),
    Compression_penalty_p50=("penalty_compression","median")
).reset_index().sort_values("RAI_mean", ascending=False)

def bucket(x):
    if x >= 70: return "Attractive"
    if x >= 40: return "Neutral"
    return "Less attractive"

summary["RAI_bucket"] = summary["RAI_mean"].apply(bucket)

# Save RAI CSVs
p_all = OUT_DIR/"role_attractiveness_index_all.csv"; summary.to_csv(p_all, index=False)
p_top = OUT_DIR/"role_attractiveness_top10.csv"; summary.head(10).to_csv(p_top, index=False)
p_bot = OUT_DIR/"role_attractiveness_bottom10.csv"; summary.tail(10).to_csv(p_bot, index=False)

# One-pager Markdown
def row_drivers(r):
    why = []
    if r.get("pct_RPP_p50",np.nan) >= 60:   why.append("strong RPP vs peers")
    if r.get("pct_RPP_p50",np.nan) <= 40:   why.append("weak RPP vs peers")
    if r.get("pct_Pay_p50",np.nan) >= 60:   why.append("pay ≥ p60 @ level")
    if r.get("pct_Pay_p50",np.nan) <= 40:   why.append("pay ≤ p40 @ level")
    if r.get("pct_PremiumSkill_p50",0) >= 30: why.append("premium-skill role")
    if r.get("Compression_penalty_p50",0) >= 60: why.append("compression risk")
    rm = r.get("Rating_median", np.nan)
    if pd.notna(rm):
        if rm >= 3.8: why.append("good rating")
        elif rm <= 3.2: why.append("rating downside")
    return ", ".join(why)

def action(rai):
    if rai >= 70: return "Stay & Stretch (high attractiveness)"
    if rai >= 40: return "Neutral: Stay with conditions / negotiate scope"
    return "Consider Negotiation / Explore Alternatives"

top10 = summary.head(10); bottom10 = summary.tail(10)
lines = ["# Role Attractiveness — One Pager", "", "## Top 10 Roles (most attractive)"]
t_title = col_title if col_title else "Job Title"
t_level = col_level if col_level else "Job Level"
t_state = col_state if col_state else "State"

for _, r in top10.iterrows():
    role = f"{r.get(t_title,'?')} | Level: {r.get(t_level,'?')} | State: {r.get(t_state,'?')}"
    drv = row_drivers(r); act = action(r["RAI_mean"])
    lines.append(f"- **{role}** — RAI {r['RAI_mean']:.1f} | RPP~{r.get('RPP_median',float('nan')):.0f} | Pay~{r.get('Pay_median',float('nan')):.0f} | {act}; Drivers: {drv}")

lines.append(""); lines.append("## Bottom 10 Roles (least attractive)")
for _, r in bottom10.iterrows():
    role = f"{r.get(t_title,'?')} | Level: {r.get(t_level,'?')} | State: {r.get(t_state,'?')}"
    drv = row_drivers(r); act = action(r["RAI_mean"])
    lines.append(f"- **{role}** — RAI {r['RAI_mean']:.1f} | RPP~{r.get('RPP_median',float('nan')):.0f} | Pay~{r.get('Pay_median',float('nan')):.0f} | {act}; Drivers: {drv}")

op = OUT_DIR/"role_attractiveness_onepager.md"
op.write_text("\n".join(lines), encoding="utf-8")

# ====================== Zip & download ======================
# Collect all outputs
to_zip = []
for p in list((OUT_DIR).glob("*.png")):
    to_zip.append(p.as_posix())
for p in list((OUT_DIR).glob("*.csv")):
    to_zip.append(p.as_posix())
to_zip.append(op.as_posix())

zip_path = OUT_DIR/"hr_insights_outputs.zip"
with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as z:
    for p in to_zip:
        if p and Path(p).exists():
            z.write(p, arcname=Path(p).name)

print(f"\nSaved outputs in: {OUT_DIR.resolve()}")
print("Files:")
for p in to_zip:
    print(" -", p)

# Try Colab download (best effort)
safe_download(zip_path)

# === Preview all charts inline (Colab/Jupyter) ===
from pathlib import Path
from IPython.display import Image, display, HTML

out_dir = Path("./hr_outputs")
pngs = sorted(out_dir.glob("*.png"))
print(f"Found {len(pngs)} chart images in {out_dir.resolve()}")

# วิธีที่ 1: โชว์ทีละรูป (อ่านชัดสุด)
for p in pngs:
    print("▶", p.name)
    display(Image(filename=str(p), width=960))  # ปรับความกว้างได้

# วิธีที่ 2 (ทางเลือก): จัดเป็นแกลเลอรี 2 คอลัมน์
cards = []
for p in pngs:
    cards.append(f"""
      <div style='flex:1 1 48%; padding:8px; box-sizing:border-box;'>
        <img src="{p.as_posix()}" style="max-width:100%; border:1px solid #ddd; border-radius:8px;">
        <div style='font:14px/1.4 monospace; margin-top:6px;'>{p.name}</div>
      </div>
    """)
html = f"<div style='display:flex; flex-wrap:wrap; gap:8px;'>{''.join(cards)}</div>"
display(HTML(html))

```

```python
# ============================================================
# FIX: Salary vs RPP bar chart with exact values (no offset)
# - Upload CSV
# - Strong numeric cleaning
# - Disable axis offset/scientific; format with commas
# - Validate sample state (e.g., "MA") before plotting
# - Plot Top 20 by avg_colid (desc) with avg_salary & avg_rpp
# ============================================================

import pandas as pd, numpy as np, re, textwrap
from pathlib import Path
import matplotlib, matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# ---------- Matplotlib config (NO OFFSET / NO SCI-NOTATION) ----------
matplotlib.rcParams.update({
    "font.size": 14,
    "axes.titlesize": 14,
    "axes.labelsize": 14,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
    "legend.fontsize": 14,
    "axes.unicode_minus": False,
    "axes.formatter.useoffset": False,  # <- สำคัญ: ปิด offset
    "axes.formatter.use_locale": False,
})

# ---------- Upload / locate CSV ----------
csv_path = None
try:
    from google.colab import files
    print("Please upload CSV (e.g., state_colid_salary_rpp.csv)")
    up = files.upload()
    for fn in up.keys():
        if fn.lower().endswith(".csv"):
            csv_path = Path(fn); break
except Exception:
    pass

if csv_path is None:
    for p in [Path("./state_colid_salary_rpp.csv"),
              Path("/content/state_colid_salary_rpp.csv"),
              Path("/mnt/data/state_colid_salary_rpp.csv")]:
        if p.exists(): csv_path = p; break

assert csv_path and csv_path.exists(), "CSV not found."
print("Using:", csv_path)

df = pd.read_csv(csv_path)

# ---------- Robust column picking ----------
def _norm(s): return re.sub(r'[^a-z0-9]+','', str(s).strip().lower())
norm_map = {_norm(c): c for c in df.columns}

def pick(*cands):
    for c in cands:
        k=_norm(c)
        if k in norm_map: return norm_map[k]
    for c in cands:
        k=_norm(c)
        for nk, orig in norm_map.items():
            if k and k in nk: return orig
    return None

col_state  = pick("state","state_code","province","abbr")
col_sal    = pick("avg_salary","avg_salary_mid","salary","average_salary","salary_mid")
col_rpp    = pick("avg_rpp","rpp","real_pay_proxy")
col_colid  = pick("avg_colid","coli","cost_of_living_index","composite index","colid")
assert col_state and col_sal, "Need state and avg_salary columns."

# ---------- Strong numeric cleaning ----------
def to_number_strong(s: pd.Series) -> pd.Series:
    def parse(x):
        if pd.isna(x): return np.nan
        t = str(x).strip().lower()
        t = re.sub(r'(usd|us\$|\$|baht|thb)', '', t)
        t = t.replace(',', '')
        t = re.sub(r'\s+', '', t)

        # range like 80k-100k
        m = re.findall(r'([-+]?[0-9]*\.?[0-9]+(?:[kmb])?)[-–]([-+]?[0-9]*\.?[0-9]+(?:[kmb])?)', t)
        if m:
            a, b = m[0]
            return np.nanmean([parse(a), parse(b)])

        # number + optional k/m/b
        m = re.fullmatch(r'([-+]?[0-9]*\.?[0-9]+)([kmb]?)', t)
        if m:
            val = float(m.group(1)); suf = m.group(2)
            if suf=='k': val *= 1e3
            elif suf=='m': val *= 1e6
            elif suf=='b': val *= 1e9
            return val

        # fallback: keep digits/.- only
        t2 = re.sub(r'[^0-9\.\-]', '', t)
        try:
            return float(t2) if t2 not in ("", ".", "-", "-.", ".-") else np.nan
        except:
            return np.nan

    return s.apply(parse)

df[col_sal] = to_number_strong(df[col_sal])
if col_rpp:
    df[col_rpp] = to_number_strong(df[col_rpp])
if col_colid:
    df[col_colid] = to_number_strong(df[col_colid])

# If no avg_rpp but have avg_colid -> compute
if (not col_rpp) or df[col_rpp].dropna().empty:
    assert col_colid, "No avg_rpp and no avg_colid to compute it."
    df["__avg_rpp_calc__"] = np.where(df[col_colid] > 0, df[col_sal] / (df[col_colid]/100.0), np.nan)
    col_rpp = "__avg_rpp_calc__"
    print("Computed avg_rpp from avg_salary and avg_colid.")

# ---------- Build working frame ----------
work = df[[col_state, col_sal, col_rpp] + ([col_colid] if col_colid else [])].copy()
work.columns = ["state", "avg_salary", "avg_rpp"] + (["avg_colid"] if col_colid else [])
work = work.replace([np.inf, -np.inf], np.nan).dropna(subset=["avg_salary","avg_rpp"])

# ---------- Quick validation for a specific state (e.g., MA) ----------
check_state = "MA"  # เปลี่ยนเป็นรัฐที่อยากเช็ค
if check_state in work["state"].astype(str).values:
    raw = df.loc[df[col_state].astype(str)==check_state, [col_sal, col_rpp] + ([col_colid] if col_colid else [])].head(3)
    used = work.loc[work["state"].astype(str)==check_state, ["avg_salary","avg_rpp"] + (["avg_colid"] if col_colid else [])].head(3)
    print(f"\n[DEBUG] {check_state} — RAW from file:")
    print(raw)
    print(f"\n[DEBUG] {check_state} — VALUES used for plot:")
    print(used)
else:
    print(f"\n[DEBUG] State '{check_state}' not found in data. Skipping debug print.")

# ---------- Pick Top 20 by avg_colid (if present), else by avg_salary ----------
if "avg_colid" in work.columns:
    top = work.sort_values("avg_colid", ascending=False).head(20)
else:
    top = work.sort_values("avg_salary", ascending=False).head(20)

# ---------- Plot (NO offset; Y shows full dollars) ----------
x = np.arange(len(top)); w = 0.42
plt.figure(figsize=(max(14, len(top)*0.8), 7))
ax = plt.gca()

# IMPORTANT: ensure plain formatting with commas, no offset
ax.ticklabel_format(style='plain', axis='y')  # <- ปิดวิทยาศาสตร์
ax.yaxis.set_major_formatter(FuncFormatter(lambda v, pos: f"{int(round(v)):,}"))

plt.bar(x - w/2, top["avg_salary"].values, width=w, label="Average Salary (Mid)")
plt.bar(x + w/2, top["avg_rpp"].values,   width=w, label="Average RPP")

plt.title("Average Salary vs RPP by State (Top 20 by COLI)")
plt.xlabel("State"); plt.ylabel("USD")

labels = [textwrap.fill(str(s), width=8) for s in top["state"].astype(str).values]
plt.xticks(x, labels, rotation=45, ha="right")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
